````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
````

````python
dataset=pd.read_csv("/content/drive/MyDrive/Dataset for research/heart attack/heart.csv")
dataset.head()
````

````python
plt.figure(figsize=(10,5))
sns.set_style('whitegrid')
sns.countplot(x='age',data=dataset)
````

````python
plt.figure(figsize=(15,6))
plt.title("Maximum heart rate achieved",fontsize=18)
sns.countplot(x='thalachh',data=dataset)
plt.show()
````

````python
plt.figure(figsize=(10,6))
sns.lineplot(x="age",y="cp",hue="output",data=dataset)
````

````python
include = dataset[dataset['output'].values == 1]
exclude = dataset[dataset['output'].values != 1]
include_1 = include
include_1
````

````python
plt.figure(figsize=(10,6))
plt.title("Heart Rate Frequency")
sns.axes_style("dark")
sns.violinplot(y=include_1["thalachh"])
plt.show()
````

````python
fig, ax = plt.subplots(2,3, figsize=(15,13))
sns.countplot(x='fbs', data=dataset, palette='magma', ax=ax[0][0]).set(title='Fasting Blood Sugar')
sns.countplot(x='exng', data=dataset, palette='magma', ax=ax[0][1]).set(title='Exercise Induced Angina')
sns.countplot(x='restecg', data=dataset, palette='magma', ax=ax[1][0]).set(title='Rest ECG')
sns.countplot(x='cp', data=dataset, palette='magma', ax=ax[0][2]).set(title='Chest Pain Type')
sns.countplot(x='caa', data=dataset, palette='magma', ax=ax[1][1]).set(title='Number of major vessels')
sns.countplot(x='thall', data=dataset, palette='magma', ax=ax[1][2]).set(title='Thallium Stress Test')
````

#Interpretations from the above graphs with categorical features.
01. More than 32% of the people have Exercise Induced Angina
47% of the people have Typical Angina or Chest Pain Type 1 and 28% have Non Anginal Pain.
02. 48% of the people have Normal Rest ECG while 50% of the people have ST-T wave abnormality.
03. More than 54% of the people have Thalium Stress test value of 2

#Graph with numeric features 

````python
fig, ax = plt.subplots(2,2, figsize=(15,13))
sns.histplot(x=dataset["age"], ax=ax[0][0], color="red", kde=True).set(title='Age')
sns.histplot(x=dataset["trtbps"], ax=ax[0][1], color="blue", kde=True).set(title='Resting Blood Pressure')
sns.histplot(x=dataset["chol"], ax=ax[1][0], color="orange", kde=True).set(title='Cholestrol Levels')
sns.histplot(x=dataset["thalachh"], ax=ax[1][1], color="green", kde=True).set(title='Maximum Heart Rate Achieved')
````

#Most of the people in this dataset have an age between 40 and 70.
01. There are a large number of people who have blood pressure in the range of 120 and 150.
02. 90/60mmHg-120/80mmHg is considered as the ideal blood pressure for a healthy person. High blood pressure is considered to be 140/90mmHg or higher.
03. Many people who have cholestrol levels in the range of 200-300mg/dL. Total cholesterol levels less than 200 milligrams per deciliter (mg/dL) are considered desirable for adults.
04. Many people have Max Heart Rate in the range of 140 and 160 bpms.According to American Heart Association the Maximum Heart Rate is around 220 beats per minute (bpm) minus the person's age.

#KDE plots

````python
fig, ax = plt.subplots(2,3, figsize=(16,13))
sns.kdeplot(x="age", data=dataset, hue="output", ax=ax[0][0], fill="True", palette="magma").set(title="Heart Attack related to Age")
sns.kdeplot(x="cp", data=dataset, hue="output", ax=ax[0][1], fill="True", palette="viridis").set(title="Heart Attack related to Chest Pain Types")
sns.kdeplot(x="thalachh", data=dataset, hue="output", ax=ax[1][0], fill="True", palette="viridis").set(title="Heart Attack related to Heart Rate")
sns.kdeplot(x="chol", data=dataset, hue="output", ax=ax[1][1], fill="True", palette="magma").set(title="Heart Attack related to Cholestrol")
sns.kdeplot(x="thall", data=dataset, hue="output", ax=ax[0][2], fill="True", palette="magma").set(title="Heart Attack related to Thallium Stress Test")
sns.kdeplot(x="trtbps", data=dataset, hue="output", ax=ax[1][2], fill="True", palette="viridis").set(title="Heart Attack related to Blood Pressure")
````

##Interpretations from the above KDE plots.
01. Most people that suffer heart attacks lie in the age of 40-60 Years
02. People that have heart attacks often suffer from Non Anginal Pain or Type 2 Chest Pain.
03. Large number of people who suffer heart attacks have Max heart rates between 140-180
04. People with Cholestrol Levels higher than 200mg/dL tend to have heart attacks more.
05. Heart Attacks were seen in people with Thallium Stress Text value of 2
